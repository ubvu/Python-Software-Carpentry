    import glob
    import numpy as np
    import matplotlib.pyplot as plt 

    path = 'data/'
    fname = 'inflammation-*.csv'
    fpaths = glob.glob(path + fname)

    sample = False

    if sample:
        fpaths = fpaths[0:3]

    fig = plt.figure(figsize=(18,9))
    for ki in np.arange(1, 13):
        filename = fpaths[ki-1]
        data = np.loadtxt(fname=filename, delimiter=',')
        axis = fig.add_subplot(3, 4, ki)
        axis.set_title('Inflammation ' + str(ki))
        axis.plot(np.mean(data, axis=0), label='mean')
        axis.plot(np.max(data, axis=0), label='max', linestyle='--')
        axis.plot(np.min(data, axis=0), label='min', linestyle=':')
        axis.set_xlim([0,40])
        if ki > 8:
             axis.set_xlabel('Day')
        if ki == 1 or ki ==5 or ki == 9:    
            axis.set_ylabel('Infection level')
        if ki == 12: 
            axis.legend()
        
    fig.tight_layout()
    plt.savefig(path + 'inflammation_overview.png')
