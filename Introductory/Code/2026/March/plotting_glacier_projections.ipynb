### A brief introduction into the glacier projections
The Open Global Glacier Model ([OGGM](https://oggm.org/)) is an open source model. It is one of the models that is being used to make projections of glacier mass loss as a result of a changing climate over the 21st century. For the <i>OGGM standard projections</i>, the model has been forced with data from climate model simulations, from different climate models and for a range of scenarios. Here we will analyse these glacier simulations ourselves.
