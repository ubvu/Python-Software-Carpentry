**Challenge 2 - Sorting Out References**

    first, second = 'Grace', 'Hopper'

**Challenge 3 - Seeing Data Types**

What are the data types of the following variables?

    planet = 'Earth'
    apples = 5
    distance = 10.5

<div class=\"alert alert-success\">
    <details>
    <summary><b></b><i>Click for a hint</i></summary>
    
Use the fuction ```type```.

<div class=\"alert alert-success\">
    <details>
    <summary><b></b><i>Click for solution</i></summary>
    
    print(type(planet))
    print(type(apples))
    print(type(distance))

<div class=\"alert alert-success\">
    <details>
    <summary><b></b><i>Click for answer</i></summary>
    
    <class 'str'>
    <class 'int'>
    <class 'float'>
