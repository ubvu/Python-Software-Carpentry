```python
def summarize_inflammation(filename, high_threshold, low_threshold=0):
    """Summarize 'suspicious' patients and days in an inflammation data file.

    A patient is 'high' if their maximum inflammation value is above high_threshold
    and their minimum value is above or equal to low_threshold.

    A day is 'always increasing' if the inflammation values for that day never go down
    when we go from the first patient to the last.
    """

    # Load data from CSV
    data = numpy.loadtxt(filename, delimiter=',')

    num_patients = len(data)
    num_days = data.shape[1]

    high_patient_ids = []          # list of patient indices
    always_increasing_days = []    # list of day indices

    # --- First loop: find "high" patients ---
    for patient in range(num_patients):
        patient_data = data[patient, :]
        patient_max = numpy.max(patient_data)
        patient_min = numpy.min(patient_data)

        if (patient_max > high_threshold) and (patient_min >= low_threshold):
            high_patient_ids.append(patient)

    # --- Second loop: find days where values never go down across patients ---
    for day in range(num_days):
        day_values = data[:, day]

        is_increasing = True
        last = day_values[0]

        for value in day_values:
            if value < last:
                is_increasing = False
            last = value

        if is_increasing:
            always_increasing_days.append(day)

    # Simple summary message
    if high_patient_ids and always_increasing_days:
        print("Found patients and days that look suspicious.")

    # ---- BUG HERE: this line raises an error ----
    suspicious_fraction = high_patient_ids / num_patients

    return high_patient_ids, always_increasing_days, suspicious_fraction


# Example call (using the SWC data file)
high_patients, inc_days, frac = summarize_inflammation("inflammation-01.csv", high_threshold=10)
print("High patients:", high_patients)
print("Always increasing days:", inc_days)
print("Fraction:", frac)
```
