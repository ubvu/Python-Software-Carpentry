## This notebook is to cover a bit about debugging

```python
# Let's make a function with an error in it to examine the message and fix it
def my_func(name, age):
    welcome_msg = "hello " + name + " how are you?"
    print(welcome_msg)
    if age is 50:
        print("congratulations, you are halfway to 100!")
        
```

```python
# Look at the error message, the first two lines are the same bu tprovide useful information
# We have a syntax error, this is like a grammatical error of the programming language
# It tells us that is is used with an int (integer) and has a suggestion
# Let's try the suggestion
def my_func(name, age):
    welcome_msg = "hello " + name + " how are you?"
    print(welcome_msg)
    if age == 50:
        print("congratulations, you are halfway to 100!")
```

```python
# No error, let's try the function out
my_func("Johnny B", 50)
```

```python
# What about if we introduce an error from the other side, the arguments?
my_func("Johnny B", "50")
```

```python
# It hasn't shown the second print out even though the age is 50
# It is not an issue with the function as it would have had a problem when defining it
# It is not an error as there is no message, but it hasn't returned what we want
# These sorts of bugs can be the most difficult to solve as there is no error message
# Let's us a print out to help us
def my_func(name, age):
    welcome_msg = "hello " + name + " how are you?"
    print(welcome_msg)
    print("the age is: ", age)
    if age == 50:
        print("congratulations, you are halfway to 100!")

my_func("Johnny B", "50")
```

```python
# Ok, so the age is being read in ok and can be displayed, what else might it be?
# We had an error earlier about using is with an integer, we are comparing the
# variable age to the integer 50, strings and integers are different, so "50" does not equal 50
# we can check this
if 50 == "50":
    print("they are the same")
else:
    print("not the same")
```

```python
# So we know they are not the same and the issue lies in providing a string instead of a number
# Note that an integer and float can be compared this way
# What happens if we ask for a comparison of 50.0 and 50 though?
# We don't need the statement to check, just claiming they are equal will perform the test too
50 == 50.0
```

```python
# Still True, that is good
```

```python
# more complex example
names = ["margaret", "seb", "jelle"]
ages = [20,62, 101]

# To loop multiple indexable variables at the same time
# we can use a comma and the zip function
for name, age in zip(names, ages):
    my_func(name, age)
```
