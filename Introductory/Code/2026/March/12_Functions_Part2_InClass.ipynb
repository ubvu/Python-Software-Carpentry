```python
# A simple function with 3 different arguments, each with a default value
def display(a=1, b=2, c=3):
    print("a: ", a, "b: ", b, "c:", c)
```

```python
# Looking at how we can use positional arguments and what the default values do
print("no parameters", display(), "one parameter", display(55))
print("two parameters", display(55, 66))
```

```python
# We can use two different named arguments, they don't need to be in the correct positional order when named
display(c=9, a=11)
```

```python
# readability of functions and code
def s(p):
    a = 0
    for v in p:
        a += v
    m = a/len(p)
    d = 0
    for v in p:
        d += (v-m) * (v-m)
    return numpy.sqrt(d/(len(p) -1))

def std_dev(sample):
    # this is a function for the standard deviation
    sample_sum = 0
    for value in sample:
        sample_sum += value

    sample_mean = sample_sum / len(sample)
    sum_squared_devs = 0
    for value in sample:
        sum_squared_devs += (value - sample_mean) * (value - sample_mean)

    return numpy.sqrt(sum_squared_devs / (len(sample) -1))

# Which one do you find more readable and why?
# It is good practice to name your functions and variables clearly
# This will help you to remember what you were doing in your code, but will help others too
# Every decision made in writing code gives context and additional information
# We want to write it in a way that anyone reading it won't have any questions
# We can do this with doc strings ("""text like this inside a function""") comments, and clear names
```

```python
# we saw that you can add ( or concatentate) strings, can you write a function that will "wrap" a
# String inside another string? such that 
# print(fence("name", "*")
# will give output
# "*name*"
def fence(string, wrapping_str):
    # We can simply add our wrapping string to either side of the main string
    # Strings can be added to each other in Python
    output_str = wrapping_str + string + wrapping_str
    return output_str

fence("hello", " yay ")
```

```python
# the str() function converts any datatype to a string
# can you write a function with 4 arguments, two of which have defaults
# that will convert all args to strings and add them together in one long string?

def stringifier(str_1, str_2, str_3="hello", str_4="world"):
    total_string = str(str_1) + str(str_2) + str(str_3) + str(str_4)
    return total_string

stringifier(str_1 = 555555, str_2 = True)
# Arguments with default values must not have arguments without defaults after
# e.g NO! func(a=1, b), YES! func(a, b=2) 
```
