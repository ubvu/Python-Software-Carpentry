```python
import numpy
data = numpy.loadtxt(fname="inflammation-01.csv", delimiter=",")
print(data)
```

```python
import matplotlib.pyplot
```

```python
image = matplotlib.pyplot.imshow(data)
# You show the plot, it is stored in working memory
matplotlib.pyplot.show() # heatmap where blue is low and yellow is high value
```

```python
# AVerage inflammation over time
avg_inflammation = numpy.mean(data, axis=0)
avg_plot = matplotlib.pyplot.plot(avg_inflammation)
matplotlib.pyplot.show()
```

```python
# maximum inflammation per day
max_plot = matplotlib.pyplot.plot(numpy.amax(data, axis=0))
matplotlib.pyplot.show()
```

```python
# minimum inflammation per day
min_plot = matplotlib.pyplot.plot(numpy.amin(data, axis=0))
matplotlib.pyplot.show()
```

```python
# We set the figure size (frame on which we place the plots) in inches
fig = matplotlib.pyplot.figure(figsize=(10, 3))
# positioning the plots
axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)
# Labelling each plot and adding it to the figure
axes1.set_ylabel('average')
axes1.plot(numpy.mean(data, axis=0))

axes2.set_ylabel('max')
axes2.plot(numpy.amax(data, axis=0))

axes3.set_ylabel('min')
axes3.plot(numpy.amin(data, axis =0))

fig.tight_layout()

# You need to save before showing to ensure that Python doesn't let it leave working memory
matplotlib.pyplot.savefig("inflammation.png")
matplotlib.pyplot.savefig("C:\\Users\\tis254\\Documents\\inflammation.png")

matplotlib.pyplot.show()
```
