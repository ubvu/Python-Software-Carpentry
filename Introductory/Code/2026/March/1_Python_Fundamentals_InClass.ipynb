This is is the intro to programming in Python. To run the cells press "shift + Enter", that will run the cell and move to the one below. If you have selected the last cell it will create a new one below. When you are not writing in the cell, but have it selected you can use 'a' to make a new cell above or 'b' to make a new one below. 

You will notice that we use the symbol "#" often, this denotes a comment. If you use it then python will ignore whatever is written after it on the same line. You can comment and uncomment lines using "ctrl + /", this will work for as many lines as you have highlighted.

```python
# Python is able to perform basic arithmetic, we use the regular + and -
# to multiply we use the "*" and to divide "/"
# You can multiply by a power by using "**"
3 + 5 * 4
```
