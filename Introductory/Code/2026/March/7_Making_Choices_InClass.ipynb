```python
# In programming languages we want our program to make decisions
# We can do this with the "if" and "else" statements
x = 10

if x == 10:
    print("success!")
# single equals is for setting a value
# double equals is for checknig truth
```

```python
# the else means that in all other situations you should do the following...
num = 37
if num > 100:
    print("greater")
else:
    print("not greater")
```

```python
# we can use elif as a second catch to check the condition
# using another if instead will break this pipeline
# the second if will start it's own new one
num = -3
if num > 0:
    print(num, "is positive")
elif num == 0:
    print(num, "is zero!")
else:
    print(num, "is negative :'(")
```

```python
# Showing the broken flow
# The first if statement is met, but the second is not
# That means that means that the else clause if triggered too
# Clearly num is not negative here
num = 2
if num > 0:
    print(num, "is positive")
if num == 0:
    print(num, "is zero!")
else:
    print(num, "is negative :'(")
```

```python
# not equal is !=
# >= greater than or equal to
# less than or equal to <=
```

```python
# We can make more complex if statements by using "and"
# Both conditions must be True to be able to move to the indent and perform the action there
# This can be used for elif statements too
if (1>0) and (-1 >= 0):
    print("both parts are true")
else:
    print("at least one part is false")
```

```python
# Another option is using "or"
# If one of the conditions is met then you will move to the indent and perform the actions there
# If both are you will move ahead to the indent, but if neither are met python will move on
if (1<0) or (1>=0):
    print("at least one is true")
```

```python
# Let's try to use some if else statements on the datasets
# We could use the statements to catch any issues
import numpy
```

```python
# First we need to load in the data and assign it to a variable
data = numpy.loadtxt(fname="inflammation-01.csv", delimiter = ",")
```

```python
# Let's make a variable for the maximum inflammation
# for the first and 21st day
# The data is inflammation of arthritic patients after trying a medicine
# so having No inflammation at the start and a peak at the point the medicine
# should have effect is a bit strange
max_inflammation_0 = numpy.amax(data, axis=0)[0]
max_inflammation_20 = numpy.amax(data, axis=0)[20]
# In the first plot we saw this strange phenomena
# In the third plot we saw that the minima on each day was always zero
# maybe a healthy person has gotten into our group, the horror!
if max_inflammation_0 == 0 and max_inflammation_20 == 20:
    print("suspicious looking maxima!")
elif numpy.sum(numpy.amin(data, axis=0)) == 0:
    print("Minima add up to zero!")
else:
    print("Seems ok!")
```

```python
# Let's check this against the third dataset now as well
data_3 = numpy.loadtxt(fname='inflammation-03.csv', delimiter=",")

max_inflammation_0 = numpy.amax(data_3, axis=0)[0]
max_inflammation_20 = numpy.amax(data_3, axis=0)[20]

if max_inflammation_0 == 0 and max_inflammation_20 == 20:
    print("suspicious looking maxima!")
elif numpy.sum(numpy.amin(data_3, axis=0)) == 0:
    print("Minima add up to zero!")
else:
    print("Seems ok!")
```
