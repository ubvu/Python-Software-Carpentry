```python
def meal_chooser():
    """
    This function opens my dataset and creates a list of the food categories to sort through 
    and offer the user a choice of the foods in the category. The choices are added to their 
    own dataframe and then shown back to the user along with the total calorie count. 
    """
    meals_df = pd.read_excel("C:/Users/tis254/Documents/Python-Software-Carpentry/Introductory/Homeworks/Gracielas_dataset.csv")
    meals_df.sort_values(by='Category')
    food_categories = list(meals_df['category'].unique())
    user_foods = pd.DataFrame(columns = meals_df.columns)
    print("Please make sure you type your choice out exactly, it is case sensitive!")
    for cat in food_categories:
        foods = list(meals_df['item'].loc[meals_df['Category'] = cat])
        user_input = ''
        while user_input not in foods:
            print("Please choose a food stuff from the following: ", foods)
            user_input = input("Type here")
        user_foods = pd.concat([user_foods, meals_df.loc[meals_df['Item'] == user_input]])
    print("Your meal consists of: " list(user_foods['Item']))
    print("Your total calorie count for your meal is: ", user_foods['Calories (per 100g)'].sum())

```

```python
# Run this cell to test the function
meal_chooser()
```
